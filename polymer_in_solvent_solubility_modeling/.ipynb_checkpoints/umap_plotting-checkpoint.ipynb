```python
import time
from pathlib import Path

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button
from rdkit import Chem
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
def move_xy(x1: list, y1: list, x2: list, y2: list, seek: float) -> (list, list):
    """Pass list of x components and y components and return list"""
    new_x = []
    new_y = []
    print(seek)
    for i in range(len(x1)):
        X = x1[i] + (x2[i] - x1[i])*seek
        Y = y2[i] + (y2[i] - y1[i])*(X - x2[i])/(x2[i] - x1[i])
        new_x.append(X)
        new_y.append(Y)
    return new_x, new_y
```

# UMAPs
This data was generated via [Uniform Manifold Approximation and Projection for Dimension Reduction (UMAP)](https://umap-learn.readthedocs.io/en/latest/). Several UMAPs were generated with a variety of parameters for two datasets. If the n_neighbors, min_dist, n_components, metric, and dtype parameters are the same, then all x and y datapoints with those equivalent parameters come from the same UMAP embedding.

UMAPs were generated from solvent or polymer fingerprints. The column headers are described as follows:
- x: x coordinate for 2d embedding
- y: y coordinate for 2d embedding
- id: unique integer used to represent polymer or solvent
- n_neighbors: value for UMAP n_neighbors parameter
- min_dist: value for UMAP min_dist parameter
- n_components: dimension of embedding to generate using UMAP. Note, this is always 2 for this dataset
- metric: metric used for UMAP metric parameter. Only cosine was used for this data as it was found to be better than euclidean or manhattan metric for molecular similarity calculations according to [literature](https://doi.org/10.1186/s13321-015-0069-3).
- dtype: fingerprint datapoint is generated from. sol_fp means it is from a solvent fingerprint embedding, pol_fp means it is from a polymer fingerprint embedding.

## Solvent UMAPs

```python
dtype = 'sol_fp'
data = pd.read_csv('sol_umaps.csv')
data = data.sort_values(by='id')
primary_alcohol = Chem.MolFromSmarts("[CH3,CH2][OH1]")
secondary_alcohol = Chem.MolFromSmarts("[CH1][OH1]")
tertiary_alcohol = Chem.MolFromSmarts("[C][OH1]")
searches = {
    #'Alkyl Halide': Chem.MolFromSmarts("[#6][F,Cl,Br,I]"),
    'Fluorine Containing': Chem.MolFromSmarts("[#6][F]"),
    'Chlorine Containing': Chem.MolFromSmarts("[#6][Cl]"),
    'Bromine Containing': Chem.MolFromSmarts("[#6][Br]"),
    'Iodine Containing': Chem.MolFromSmarts("[#6][I]"),
    'Ether': Chem.MolFromSmarts("[OD2]([#6])[#6]"),
    'Amine': Chem.MolFromSmarts("[NX3;H2,H1;!$(NC=O)]"),
    'Benzene Containing': Chem.MolFromSmarts("C1=CC=CC=C1"),
    'Sulfur Containing': Chem.MolFromSmarts("[S]"),
    'Aldehyde': Chem.MolFromSmarts("[CX3H1](=O)[#6]"),
    'Amide': Chem.MolFromSmarts("[NX3][CX3](=[OX1])[#6]"),
    'Acid': Chem.MolFromSmarts("[CX3,NX3,SX4](=O)[OX2H1]"),
    'Ketone': Chem.MolFromSmarts("[#6][CX3](=O)[#6]"),
    'Nitrile': Chem.MolFromSmarts("[NX1]#[CX2]"),
    'Phosphorus Containing': Chem.MolFromSmarts("[P]"),
}
# need to negate anhydride
ester = Chem.MolFromSmarts("[#6][CX3](=O)[OX2H0][#6]")
anhydride = Chem.MolFromSmarts("[CX3](=[OX1])[OX2][CX3](=[OX1])")

hue = []
for index, row in data.iterrows():
    smiles = row['smiles']
    m = Chem.MolFromSmiles(smiles)
    N_in = False
    hal_in = False
    class_ = ""
    if (len(m.GetSubstructMatches(primary_alcohol)) > 0 or
            len(m.GetSubstructMatches(secondary_alcohol)) > 0 or
            len(m.GetSubstructMatches(tertiary_alcohol)) > 0):
        class_ += 'Alcohol'
    for k, v in searches.items():
        if len(m.GetSubstructMatches(v)) > 0:
            if len(class_) > 0:
                class_ += f", {k}"
            else:
                class_ += f"{k}"

    if (len(m.GetSubstructMatches(ester)) > 0 and 
          len(m.GetSubstructMatches(anhydride)) == 0):
        if len(class_) > 0:
            class_ += f", Ester"
        else:
            class_ += f"Ester"
    for atom in m.GetAtoms():
        if atom.GetIsAromatic():
            if len(class_) > 0:
                class_ += f", Aromatic"
            else:
                class_ += "Aromatic"
            break
    
    if len(class_) == 0:
        class_ = 'Only Alkane/Alkene'
    if row['name'] == 'DMF' or row['name'] == 'formamide':
        class_ = 'Amide'
    if row['smiles'] == 'Cl':
        class_ = 'Chlorine Containing'
    if row['name'] == 'oxidane':
        class_ = 'Water'
    if class_ == 'Ether, Ester':
        class_ = 'Ester'
    if class_ == 'Alcohol, Acid':
        class_ = 'Acid'
    if class_ == 'Alcohol, Chlorine Containing, Acid':
        class_ = 'Chlorine Containing, Acid'
    if class_ == 'Alcohol, Fluorine Containing, Acid':
        class_ = 'Fluorine Containing, Acid'
    hue.append(class_)

     
data['Functional Groups'] = hue
order = list(data['Functional Groups'].unique())
order.sort()
```

```python
plt.close('all')
steps = 40
plot_margin = 1
valmin_min_dist = min(data.min_dist)
valmax_min_dist = max(data.min_dist)
valstep_min_dist = list(data.min_dist.unique())
valmin_n_neighbors = min(data.n_neighbors)
valmax_n_neighbors = max(data.n_neighbors)
valstep_n_neighbors = list(data.n_neighbors.unique())


plot_df = data.loc[
      (data.n_neighbors == 3)
    & (data.min_dist == 0)
]
fig, ax = plt.subplots(figsize=(12,9))
sc = sns.scatterplot(x='x', y='y', hue='Functional Groups', style='Functional Groups', s=100, data=plot_df, 
                hue_order=order, ax=ax)
ax.legend(bbox_to_anchor=(1, 1))
plt.tight_layout()
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('')
ax.set_ylabel('')
ax.set_xlim([min(data.x) - plot_margin, max(data.x) + plot_margin])
ax.set_ylim([min(data.y) - plot_margin, max(data.y) + plot_margin])

# adjust the main plot to make room for the sliders
plt.subplots_adjust(left=0.25, bottom=0.25)
ax_mdist = plt.axes([0.25, 0.1, 0.65, 0.03])
mdist_slider = Slider(
    ax=ax_mdist,
    label='min_dist',
    valmin=valmin_min_dist,
    valmax=valmax_min_dist,
    valinit=valmin_min_dist,
    valstep=valstep_min_dist,
)
ax_nneigh = plt.axes([0.25, .05, 0.65, 0.03])
nneigh_slider = Slider(
    ax=ax_nneigh,
    label='n_neighbors',
    valmin=valmin_n_neighbors,
    valmax=valmax_n_neighbors,
    valinit=valmin_n_neighbors,
    valstep=valstep_n_neighbors,
)

# The function to be called anytime a slider's value changes
def update(val):
    old_x = plot_df.x.to_list()
    old_y = plot_df.y.to_list()
    plot_df = data.loc[
          (data.n_neighbors == nneigh_slider.val)
        & (data.min_dist == mdist_slider.val)
    ]
    new_x = plot_df.x.to_list()
    new_y = plot_df.y.to_list()
    functional_groups = plot_df['Functional Groups']
    for i in range(steps):
        ax.clear()
        x, y = move_xy(old_x, old_y, new_x, new_y, (i+1)/steps)
        sc = sns.scatterplot(x=x, y=y, hue=functional_groups, style=functional_groups, s=100, data=plot_df, 
                hue_order = order, ax=ax)
    ax.clear()
    sc = sns.scatterplot(x=x, y=y, hue=functional_groups, style=functional_groups, s=100, data=plot_df, 
                hue_order=order, ax=ax)

mdist_slider.on_changed(update)
nneigh_slider.on_changed(update)
plt.show()
```
